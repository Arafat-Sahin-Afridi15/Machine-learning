```python
import pandas as pd

data = pd.read_csv("./titanic.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
data2 = data
data2['Age_mean'] = data2['Age'].fillna(data2['Age'].mean())
```

```python
data2.isna().sum()
```

```python
import matplotlib.pyplot as plt

data2[['Age','Age_mean']].plot(kind='kde')
plt.show()
```

```python
random_selection = data2['Age'].dropna().sample(data2['Age'].isna().sum())
```

```python
random_selection.index = data2['Age'].loc[data2['Age'].isna()].index
```

```python
data['Age_random']=data2['Age'].fillna(random_selection)
```

```python
import matplotlib.pyplot as plt

data2[['Age','Age_random']].plot(kind='kde')
plt.show()
```
